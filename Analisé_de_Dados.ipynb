# Ínicio


* Importando biblioteca Pandas;
* Importando uma base de dados pelo Googlo Colab;

```python
import pandas as pd
```

```python
# "read_excel" -> Lê o conteúdo da planilha "houses_rent.xlsx"
# aluguel_casas -> Recebe o conteúdo da planilha "houses_rent.xlsx"

aluguel_casas = pd.read_excel("/content/houses_rent.xlsx")
```

# Pré-processamento de Dados


* Padronização de cabeçalhos: renomeação e remoção de espaços em branco;

* Análise e inspeção: de tipos de dados das colunas;

* Conversão de tipos de dados: ajuste de colunas numéricas para int ou float.

```python
# "head()" -> moastra as 5 primeiras linhas do DataFrame.
aluguel_casas.head()
```

```python
# "tail()" -> mostra as 5 últimas linhas do DataFrame.
aluguel_casas.tail()
```

```python
# "rename(columns = {})" -> Renomea o cabeçalho, removendo os espaços em branco.
#  "aluguel_casas" -> Recebe a modificação do cabeçalho.
aluguel_casas = aluguel_casas.rename(columns = {"parking spaces": "parking_spaces", "rent amount": "rent_amount", "property tax": "property_tax", "fire insurance": "fire_insurance"})
```

```python
# ".dtypes" -> verifica o tipo dos objeto/objetos em um dataframe
# "()" -> se utilizar no final vai apresentar um erro "'Series' object is not callable"
aluguel_casas.dtypes
```

```python
# ".astype" -> altera o tipo de dados
aluguel_casas = aluguel_casas.astype({"rent_amount": "float", "property_tax": "float", "fire_insurance": "float", "total": "float"})

# Mostra o valor alterado
aluguel_casas.dtypes
```

# Análise Estrutural do DataFrame


A função .info() do Pandas fornece um resumo da estrutura do DataFrame, exibindo:

* Número total de entradas e índice associado;

* Nome das colunas e quantidade de valores não nulos em cada uma;

* Tipo de dado (dtype) de cada coluna;

* Uso de memória do DataFrame.

Essa função é útil para identificar valores ausentes, verificar tipos de dados e otimizar o consumo de memória ao trabalhar com grandes conjuntos de dados.

```python
# ".info()" -> resumo do DataFrame
aluguel_casas.info()
```

# Introdução a Estatística


A função .describe() do Pandas gera estatísticas descritivas de um DataFrame ou de uma Série. Ela fornece um resumo estatístico das colunas numéricas por padrão, mas pode ser usada também para colunas categóricas quando especificado.

Por padrão, .describe() retorna as seguintes estatísticas para colunas
numéricas (int64, float64):

* count: Número de valores não nulos
* mean: Média aritmética
* std: Desvio padrão
* min: Menor valor
* 25%: Primeiro quartil (Q1)
* 50%: Mediana (Q2)
* 75%: Terceiro quartil (Q3)
* max: Maior valor

Se aplicado a colunas categóricas (object, category), retorna:

* count: Número de valores não nulos
* unique: Número de valores únicos
* top: Valor mais frequente
* freq: Frequência do valor mais frequente

```python
# Estatística Descritiva para colunas numéricas.
# Seleciona as colunas com o padrão "DataFrame[["coluna1", "coluna2", "coluna3", ... ]]".
# ".describe()" -> Retorna as estatísticas descritivas do DataFrame.
aluguel_casas[["area", "rooms", "bathroom", "parking_spaces", "rent_amount", "property_tax", "fire_insurance", "total"]].describe()
```

```python
# Estatísticas Descritivas para colunas categóricas.
# "include =["object"] -> "Seleciona as colunas, todas, "object".
# ".describe()" -> Retorna as estatísticas descritivas do DataFrame.
aluguel_casas.describe(include=['object'])
```
